```python
# Tokenizerë¥¼ í†µí•´ ìì—°ì–´->ë²¡í„° Encoding
input_text = "10ë…„ ë…¼ë€ ëì— ë°‘ê¸€ì”¨ê¹Œì§€ ìƒˆê¸°ê³  ì œì‘ ì™„ë£Œë¥¼ ëˆˆì•ì— ë‘” â€˜ê´‘í™”ë¬¸ í˜„íŒâ€™ì„ ì›ì ì—ì„œ ì¬ê²€í† í•´ í•œê¸€ í˜„íŒìœ¼ë¡œ êµì²´í•˜ìëŠ” ì£¼ì¥ì´ ë¬¸í™”ê³„ ëª…ì‚¬ë“¤ì´ í¬í•¨ëœ ìì¹­ â€˜ì‹œë¯¼ëª¨ì„â€™ì—ì„œ ë‚˜ì™”ë‹¤.\n  ì´ë“¤ì€ ë¬¸í™”ì¬ì²­ì´ ì§€ë‚œí•´ 8ì›” ìµœì¢… í™•ì •í•œ ë³µì›ì•ˆì´ ì‹œë¯¼ ì˜ê²¬ê³¼ ì‹œëŒ€ì •ì‹ ì„ ë°˜ì˜í•œ ê²ƒì´ ì•„ë‹ˆë¼ë©´ì„œ ì˜¤ëŠ” í•œê¸€ë‚ ê¹Œì§€ ëŒ€ëŒ€ì ì¸ í˜„íŒ êµì²´ ì‹œë¯¼ìš´ë™ì„ ë²Œì´ê² ë‹¤ê³  ì˜ˆê³ í–ˆë‹¤.\n  â€˜ê´‘í™”ë¬¸ í˜„íŒ í›ˆë¯¼ì •ìŒì²´ë¡œ ì‹œë¯¼ëª¨ì„â€™(ê³µë™ëŒ€í‘œ ê°•ë³‘ì¸â€§í•œì¬ì¤€, ì´í•˜ â€˜ì‹œë¯¼ëª¨ì„â€™)ì— ì´ë¦„ì„ ì˜¬ë¦° ë¬¸í™”ì˜ˆìˆ ì¸ì€ í˜„ì¬ê¹Œì§€ ì´ 24ëª….\n  ì´ ì¤‘ì—” 2014~2016ë…„ ì„œìš¸ì‹œ ì´ê´„ê±´ì¶•ê°€ë¥¼ ì§€ë‚¸ ìŠ¹íš¨ìƒ ì´ë¡œì¬ ëŒ€í‘œì™€ â€˜ì•ˆìƒìˆ˜ì²´â€™ë¡œ ìœ ëª…í•œ ì•ˆìƒìˆ˜ íŒŒì£¼íƒ€ì´í¬ê·¸ë¼í”¼í•™êµ êµì¥, ìœ ì˜ìˆ™ ì „ í™˜ê²½ë¶€ì¥ê´€(ì„¸ì¢…ì‚¬ë‘ë°© íšŒì¥), ì„ì˜¥ìƒ ë¯¸ìˆ ê°€ ë“±ì´ ìˆë‹¤.\n  ê³µë™ëŒ€í‘œì¸ ê°•ë³‘ì¸ ì‘ê°€ëŠ” â€˜ì°¸ì´ìŠ¬â€™ â€˜í™”ìš”â€™ ë“±ì˜ ìƒí‘œ ê¸€ì”¨ë¡œ ìœ ëª…í•œ ìº˜ë¦¬ê·¸ë¼í”¼(ì„œì²´) ì‘ê°€ë‹¤.\n  â€˜ì‹œë¯¼ëª¨ì„â€™ì€ 14ì¼ ì˜¤í›„ ì„œìš¸ ì¢…ë¡œêµ¬ì˜ í•œ ì„œì ì—ì„œ ê¸°ìê°„ë‹´íšŒë¥¼ ì—´ê³  ì´ ê°™ì€ ì…ì¥ê³¼ í•¨ê»˜ í›ˆë¯¼ì •ìŒ í•´ë¡€ ê¸€ìê¼´ë¡œ ì‹œë²” ì œì‘í•œ ëª¨í˜• í˜„íŒ(1/2 í¬ê¸° ì¶•ì†ŒíŒ)ë„ ê³µê°œí•  ì˜ˆì •ì´ë‹¤.\n  ê°• ê³µë™ëŒ€í‘œëŠ” 13ì¼ ê¸°ìì™€ í†µí™”ì—ì„œ â€œìƒˆ í˜„íŒ ì œì‘ ê³¼ì •ì—ì„œ í•œê¸€ë¡œ ë§Œë“¤ìëŠ” ì˜ê²¬ì€ ë¬µì‚´ëë‹¤â€ë©´ì„œ â€œì§€ë‚œí•´ 8ì›” ì´í›„ ë¬¸í™”ì¬ì²­ì— ê±°ë“­ ì…ì¥ì„ ì „í–ˆì§€ë§Œ ë°˜ì˜ë˜ì§€ ì•Šì•„ ì‹œë¯¼ìš´ë™ì— ë‚˜ì„œê¸°ë¡œ í–ˆë‹¤â€ê³  ë§í–ˆë‹¤.\n  ì¼ë‹¨ ë¬¸í™”ì˜ˆìˆ ì¸ ì£¼ì¶•ìœ¼ë¡œ ê¾¸ë ¸ì§€ë§Œ ì¡°ë§Œê°„ í•œê¸€í˜‘íšŒ ë“± í•œê¸€ ê´€ë ¨ë‹¨ì²´ë“¤ê³¼ ì—°ëŒ€í•œë‹¤ëŠ” ë°©ì¹¨ì´ë‹¤.\n  ì´ë“¤ì´ ë°°í¬í•œ ì‚¬ì „ìë£Œì—” ^í•œìí˜„íŒ ì„¤ì¹˜ëŠ” ì¤‘êµ­ì˜ ì†êµ­ì„ì„ í‘œì‹œí•˜ëŠ” ê²ƒìœ¼ë¡œ ëŒ€í•œë¯¼êµ­ ì •ì²´ì„±ì— ë„ì›€ì´ ë˜ì§€ ì•Šê³  ^ê´‘í™”ë¬¸ì€ 21ì„¸ê¸°ì˜ ì¤‘ê±´ì´ì§€ ë³µì›ì´ ì•„ë‹ˆë¯€ë¡œ ë‹¹ëŒ€ì˜ ì‹œëŒ€ì •ì‹ ì¸ í•œê¸€ë¡œ í˜„íŒì„ ì¨ì•¼í•˜ë©° ^í•œê¸€í˜„íŒì€ ë¯¸ë˜ì— ë‚¨ê²¨ì¤„ ìš°ë¦¬ ìœ ì‚°ì„ ì¬ì°½ì¡°í•œë‹¤ëŠ” ì˜ë¯¸ë¼ëŠ” ì£¼ì¥ì´ ë‹´ê²¼ë‹¤.\n  í˜„ì¬ ê´‘í™”ë¬¸ í˜„íŒì— ëŒ€í•´ì„  â€œê³ ì¢…ì´ ê²½ë³µê¶ì„ ì¤‘ê±´í•  ë•Œ ë‹¹ì‹œ í›ˆë ¨ëŒ€ì¥ì´ë˜ ì„íƒœì˜ì´ ì“´ ê´‘í™”ë¬¸ í˜„íŒì˜ ê¸€ì”¨ë¥¼ ì¡°ê·¸ë§Œ ì‚¬ì§„ì—ì„œ ìŠ¤ìºë‹í•˜ê³  ì´ë¥¼ ë‹¤ë“¬ì–´ ì´ëª…ë°•ì •ë¶€ ë•Œ ì„¤ì¹˜ëœ ê²ƒâ€ì´ë¼ë©´ì„œ ë³µì› ê¸°ì¤€ìœ¼ë¡œì„œì˜ ì •ë‹¹ì„±ì„ ê¹ì•„ë‚´ë ¸ë‹¤.\n    â€˜ì‹œë¯¼ëª¨ì„â€™ì— ì°¸ì—¬í•œ ìŠ¹íš¨ìƒ ëŒ€í‘œë„ ê°œì¸ì˜ê²¬ì„ ì „ì œë¡œ â€œí˜„íŒì„ ê¼­ í•œê°€ì§€ë§Œ ê³ ì§‘í•  í•„ìš”ë„ ì—†ë‹¤.\n  ë§¤ë…„ êµì²´í•  ìˆ˜ë„ ìˆê³ , ê´‘ì¥ì—ì„œ ë³´ì´ëŠ” ì •ë©´ì—” í•œê¸€í˜„íŒ, ë°˜ëŒ€í¸ì—” í•œìí˜„íŒì„ ë‹¤ëŠ” ì•„ì´ë””ì–´ë„ ê°€ëŠ¥í•œ ê²ƒ ì•„ë‹ˆëƒâ€ê³  ë§í–ˆë‹¤.\n  ê·¸ëŸ¬ë©´ì„œ â€œë¬¸í™”ì¬ ì „ë¬¸ê°€ë“¤ì€ ë³´ìˆ˜ì ì¼ ìˆ˜ë°–ì— ì—†ì§€ë§Œ í˜„íŒì´ë€ ê²Œ ìš”ì¦˜ ë§ë¡œëŠ” â€˜ê°„íŒâ€™ì¸ë° ìƒˆ ì‹œëŒ€ì— ë§ê²Œ ë°”ê¿” ë‹¤ëŠ” ê²Œ ë°”ëŒì§í•˜ë‹¤â€ê³  ì£¼ì¥í–ˆë‹¤.\n"
inputs = tokenizer(input_text, return_tensors="pt", padding="max_length", truncation=True, max_length=1026) # ë§¤ê°œë³€ìˆ˜ë“¤ì˜ ì˜ë¯¸ëŠ” ì§€ê¸ˆ ì•ˆì¤‘ìš”í•¨
print(inputs)
```

1ë²ˆ ì½”ë“œë¸”ëŸ­ì—ì„œ ìš°ë¦¬ëŠ” *model*ë³€ìˆ˜ì— *EbanLee/kobart-summary-v3* ëª¨ë¸ì„ ë‹´ì•˜ìŠµë‹ˆë‹¤. ì•„ë˜ì˜ ì½”ë“œì—ì„œëŠ” ì´ëŸ¬í•œ *model*ë³€ìˆ˜ì˜ *model.generate*í•¨ìˆ˜ë¥¼ í†µí•´ kobartëª¨ë¸ì—ê²Œ ì…ë ¥ì„ ì£¼ê³  ê·¸ì— í•´ë‹¹í•˜ëŠ” ì¶œë ¥ì„ ë°›ì•„ë´…ë‹ˆë‹¤. <br><br>
model.generateí•¨ìˆ˜ëŠ” input_idsì™€ attention_maskë¥¼ ì…ë ¥ìœ¼ë¡œ ë°›ê³ , ë²¡í„°í˜•íƒœì˜ ê²°ê³¼ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.

```python
# ë³€í™˜í•œ ë‘ ë²¡í„°ë¥¼ kobartì—ê²Œ ì…ë ¥ê°’ìœ¼ë¡œ ì£¼ê¸°
summary_text_ids = model.generate(
input_ids=inputs['input_ids'], # ìì—°ì–´ë¥¼ tokenizerë¡œ ë³€í™˜í•œ ê²°ê³¼ 1
attention_mask=inputs['attention_mask'], # ìì—°ì–´ë¥¼ tokenizerë¡œ ë³€í™˜í•œ ê²°ê³¼ 2
bos_token_id=model.config.bos_token_id,
eos_token_id=model.config.eos_token_id,
length_penalty=1.0,
max_length=300,
min_length=12,
num_beams=6,
repetition_penalty=1.5,
no_repeat_ngram_size=15,
)
# kobartì˜ ì¶œë ¥ì„ printí•´ë³´ê¸° (ë„ˆë¬´ ê¸¸ê¸° ë•Œë¬¸ì— ì²« 100ê°œë§Œ ì¶œë ¥í•˜ê²Œí•¨)
print(summary_text_ids[0][:100])
```

```python
# kobartì˜ ì¶œë ¥ì„ tokenizerë¡œ ë‹¤ì‹œ decodingí•´ì„œ ìì—°ì–´ë¡œ ë°”ê¾¸ê¸°
result = tokenizer.decode(summary_text_ids[0], skip_special_tokens=True) #tokenizer.decodeë¥¼ í†µí•´ ë²¡í„°ë¥¼ ìì—°ì–´ë¡œ ë³€í™˜
print(result)
```

## 2.3 í•™ìŠµì— í•„ìš”í•œ ë‹¤ì–‘í•œ í•¨ìˆ˜&classì •ì˜
**ì´í•´í•  í•„ìš” ì—†ìŒ**. (ì–´ì°¨í”¼ ì´ë²ˆ í•™ê¸°ì— ì¸ê³µì§€ëŠ¥í•˜ê³  ê¸°ê³„í•™ìŠµ ê°•ì˜ì—ì„œ ë‹¤ ë°°ìš¸ ë‚´ìš©ë“¤ì„. ì§€ê¸ˆì€ ë¬´ì‹œë¬´ì‹œí•´ë³´ì—¬ë„ ì´ë²ˆí•™ê¸° ëë‚˜ë©´ ë‹¤ ë‹¹ì—°í•œ ë‚´ìš©ë“¤ì´ ë ê±°ì„)ê·¸ë˜ë„ ê°„ëµí•˜ê²Œ ì„¤ëª…í•˜ìë©´

*   *preprocess_data*í•¨ìˆ˜ëŠ” tokenizerë¥¼ í†µí•´ ë°ì´í„°ë¥¼ ë²¡í„°ì˜ í˜•íƒœë¡œ ë³€í™˜í•˜ëŠ” ì „ì²˜ë¦¬ê³¼ì •ì„ ê±°ì¹˜ëŠ” í•¨ìˆ˜ì´ê³ ,
*   *SummaryDataset(Dataset)* ëŠ” ì „ì²˜ë¦¬ë¥¼ ê±°ì¹œ ë²¡í„° í˜•íƒœì˜ ë°ì´í„°ë“¤ì„ ì‚¬ìš©í•˜ê¸° í¸í•˜ê²Œ ë¬¶ì–´ë‘” ê°ì²´ì´ê³ ,
*   *train_dataset*ëŠ” SummaryDataset(Dataset)ë¥¼ ë³€ìˆ˜ë¡œ í• ë‹¹í•œ ê²ƒì´ê³ ,
*   *train_dataloader*ëŠ” í•™ìŠµ ëŠ¥ë¥ ì„ ìœ„í•´ì„œ ë°ì´í„°ë¥¼ 4ê°œì˜ [mini-batchë¡œ](https://wikidocs.net/55580) ìª¼ê°œê³  ìˆœì„œë¥¼ ì„ì–´ì£¼ëŠ” í•¨ìˆ˜ì´ë‹¤ (ê¶ê¸ˆí•˜ë©´ ë“¤ì–´ê°€ë´)
*   *optimizer*ëŠ” í•™ìŠµí• ë•Œ ì„¤ì •í•˜ëŠ” ë§¤ê°œë³€ìˆ˜ë“¤ì„ ìµœì í™”í•˜ëŠ” í•¨ìˆ˜ì´ë‹¤. [optimizerê³µë¶€](https://wikidocs.net/55580)(ì—­ì‹œ, ê¶ê¸ˆí•˜ë©´ ë“¤ì–´ê°€ë´)



```python
# 3. ë°ì´í„° ì „ì²˜ë¦¬ í•¨ìˆ˜ ì •ì˜
def preprocess_data(tokenizer, data, max_input_length=1024, max_target_length=128):
    inputs = []
    targets = []

    for input_text, summary_text in data:
        # ì…ë ¥ í…ìŠ¤íŠ¸ ì „ì²˜ë¦¬
        input_encodings = tokenizer(
            input_text,
            max_length=max_input_length,
            padding="max_length",
            truncation=True,
            return_tensors="pt"
        )

        # ìš”ì•½ í…ìŠ¤íŠ¸ ì „ì²˜ë¦¬
        target_encodings = tokenizer(
            summary_text,
            max_length=max_target_length,
            padding="max_length",
            truncation=True,
            return_tensors="pt"
        )

        inputs.append(input_encodings)
        targets.append(target_encodings)

    return inputs, targets

inputs, targets = preprocess_data(tokenizer, train_data)

# 4. ì»¤ìŠ¤í…€ ë°ì´í„°ì…‹ ì •ì˜
class SummaryDataset(Dataset):
    def __init__(self, inputs, targets):
        self.inputs = inputs
        self.targets = targets

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):
        input_ids = self.inputs[idx]['input_ids'].squeeze()
        attention_mask = self.inputs[idx]['attention_mask'].squeeze()
        labels = self.targets[idx]['input_ids'].squeeze()

        return {
            "input_ids": input_ids,
            "attention_mask": attention_mask,
            "labels": labels,
        }

# 5. ë°ì´í„°ë¡œë” ìƒì„±
train_dataset = SummaryDataset(inputs, targets)
train_dataloader = DataLoader(train_dataset, batch_size=4, shuffle=True)

# 6. ì˜µí‹°ë§ˆì´ì € ì„¤ì •
optimizer = AdamW(model.parameters(), lr=5e-5)
```

## 2.4 ë³¸ê²©ì ìœ¼ë¡œ ëª¨ë¸ í•™ìŠµì‹œì¼œë³´ê¸° (ì˜¤ë˜ê±¸ë¦¼)
**num_epochs**ëŠ” ìš°ë¦¬ê°€ ìš°ë¦¬ì˜ ë°ì´í„°ë¥¼ ëª‡ë²ˆ í•™ìŠµ ì‹œí‚¬ê²ƒì¸ì§€. (í˜„ì¬ëŠ” 4ì´ë¯€ë¡œ ìš°ë¦¬ì˜ ë°ì´í„°ë¥¼ 4ë²ˆ í•™ìŠµì‹œí‚´)<br><br><br> **í•™ìŠµ ê³¼ì • ì„¤ëª…**


1.   ì…ë ¥ ë°ì´í„°ë¥¼ í†µí•´ í˜„ì¬ ìš°ë¦¬ì˜ ëª¨ë¸ì˜ ì˜ˆì¸¡ì„ ì–»ìŒ(outputs)
2.   í˜„ì¬ ìš°ë¦¬ ëª¨ë¸ì˜ ì˜ˆì¸¡ì´ ì •ë‹µê³¼ ì–¼ë§ˆë‚˜ ë‹¤ë¥¸ì§€ ê³„ì‚°. ì˜ˆì¸¡ê³¼ ì •ë‹µì˜ ì°¨ì´ëŠ” lossê°’ì„. ì‘ì„ìˆ˜ë¡ ì¢‹ì€ê²ƒì´ë‹¤.
3. lossê°’ì— ë”°ë¥¸ *ì˜¤ì°¨ì—­ì „íŒŒ*(ì´ê²ƒë„ ì´ë²ˆí•™ê¸° ì´ˆë°˜ì— ë°°ìš¸ê²ƒì„)ë¥¼ í†µí•´ ìš°ë¦¬ì˜ ëª¨ë¸ì„ í•™ìŠµì‹œí‚´
4. ì´ ê³¼ì •ì„ epoch(4ë²ˆ)ë§Œí¼ ë°˜ë³µ (epochê°€ ì§„í–‰ë¨ì— ë”°ë¼ lossê°€ ì¤„ì–´ë“œëŠ” ê²ƒì„ ë³¼ ìˆ˜ ìˆë‹¤. ì´ê²ƒì€ ìš°ë¦¬ ëª¨ë¸ì´ í•™ìŠµí•˜ë©´ì„œ ì •ë‹µê³¼ ì˜ˆì¸¡ ì‚¬ì´ì˜ ì˜¤ì°¨ê°€ ì¤„ì–´ë“¤ê³  ìˆë‹¤ëŠ” ê²ƒì´ë‹¤)





```python
# 7. ëª¨ë¸ í•™ìŠµ
model.train()
num_epochs = 4 # ë°ì´í„°ë¥¼ ëª‡ë²ˆ í•™ìŠµì‹œí‚¬ ê²ƒì¸ì§€

for epoch in range(num_epochs): # ì´ 3ë²ˆ ë°˜ë³µ
    for batch in train_dataloader: # train_dataloaderëŠ” ìš°ë¦¬ì˜ ë°ì´í„°ë¥¼ 4ê°œë¡œ ì‘ê²Œ ìª¼ê°œë†“ì€ ë°°ì—´
        optimizer.zero_grad() # ì•ˆì¤‘ìš”. ê·¸ëƒ¥ ë‹¤ì–‘í•œ ë§¤ê°œë³€ìˆ˜ë¥¼ ìµœì í™”

        # ë°°ì¹˜ ë°ì´í„° ì¶”ì¶œ
        input_ids = batch["input_ids"] # kobartê°€ í•„ìš”ë¡œí•˜ëŠ” ì²«ë²ˆì§¸ ë²¡í„° input_id
        attention_mask = batch["attention_mask"] # kobartê°€ í•„ìš”ë¡œí•˜ëŠ” ì²«ë²ˆì§¸ ë²¡í„° attention_mask
        labels = batch["labels"] # ìš°ë¦¬ì˜ ë°ì´í„°ì—ì„œ ì •ë‹µê°’ì„ tokenizerë¥¼ í†µí•´ ë²¡í„°ë¡œ ë°”ê¿”ë‘” ê²ƒ ì¤‘ input_ids

        # ëª¨ë¸ ì˜ˆì¸¡
        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            labels=labels
        )

        # ì†ì‹¤ ê³„ì‚° ë° ì—­ì „íŒŒ
        loss = outputs.loss
        loss.backward()
        optimizer.step()

    print(f"Epoch {epoch + 1} completed. Loss: {loss.item()}")
```

# 3. kobart ì‚¬ìš© ì½”ë“œ (í”„ë¡¬í”„íŠ¸ ì—”ì§€ë‹ˆì–´ë§ ì‹¤ìŠµìš©)

```python
from transformers import PreTrainedTokenizerFast, BartForConditionalGeneration

# í† í¬ë‚˜ì´ì €ì™€ ëª¨ë¸ì„ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.
tokenizer = PreTrainedTokenizerFast.from_pretrained("EbanLee/kobart-summary-v3")
model = BartForConditionalGeneration.from_pretrained("EbanLee/kobart-summary-v3")

# Tokenizerë¥¼ í†µí•´ ìì—°ì–´->ë²¡í„° Encoding
input_text = "ì—¬ê¸°ì— ë„¤ê°€ ì¡°ë¦½í•œ í”„ë¡¬í”„íŠ¸ë¥¼ ë„£ìœ¼ë©´ë©ë‹ˆë‹¤."
inputs = tokenizer(input_text, return_tensors="pt", padding="max_length", truncation=True, max_length=1026) # ë§¤ê°œë³€ìˆ˜ë“¤ì˜ ì˜ë¯¸ëŠ” ì§€ê¸ˆ ì•ˆì¤‘ìš”í•¨
print(inputs)

# ë³€í™˜í•œ ë‘ ë²¡í„°ë¥¼ kobartì—ê²Œ ì…ë ¥ê°’ìœ¼ë¡œ ì£¼ê¸°
summary_text_ids = model.generate(
input_ids=inputs['input_ids'], # ìì—°ì–´ë¥¼ tokenizerë¡œ ë³€í™˜í•œ ê²°ê³¼ 1
attention_mask=inputs['attention_mask'], # ìì—°ì–´ë¥¼ tokenizerë¡œ ë³€í™˜í•œ ê²°ê³¼ 2
bos_token_id=model.config.bos_token_id,
eos_token_id=model.config.eos_token_id,
length_penalty=1.0,
max_length=300,
min_length=12,
num_beams=6,
repetition_penalty=1.5,
no_repeat_ngram_size=15,
)

# kobartì˜ ì¶œë ¥ì„ tokenizerë¡œ ë‹¤ì‹œ decodingí•´ì„œ ìì—°ì–´ë¡œ ë°”ê¾¸ê¸°
result = tokenizer.decode(summary_text_ids[0], skip_special_tokens=True) #tokenizer.decodeë¥¼ í†µí•´ ë²¡í„°ë¥¼ ìì—°ì–´ë¡œ ë³€í™˜
print(result)
```
